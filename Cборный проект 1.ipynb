- –í—ã—Ä—É—á–∫–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –æ–±—ä–µ–º–∞ –≥–æ—Å –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–π

```python
# –§–∏–ª—å—Ç—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ: —Ç–æ–ª—å–∫–æ —Ñ–∏–ª—å–º—ã —Å –≥–æ—Å. –ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –∏ –Ω–µ–Ω—É–ª–µ–≤—ã–º–∏ —Å–±–æ—Ä–∞–º–∏
filtered_df = df_gos.loc[
    (df_gos['refundable_support'] + df_gos['nonrefundable_support'] > 0) & 
    (df_gos['box_office'].notna() & df_gos['box_office'] > 0)
]
#filtered_df
```

```python
# –°—á–∏—Ç–∞–µ–º –æ–±—â—É—é –ø–æ–¥–¥–µ—Ä–∂–∫—É, —Å–±–æ—Ä—ã –∏ —Å—Ä–µ–¥–Ω–∏–π ROS –ø–æ —Ä–µ–∂–∏—Å—Å–µ—Ä–∞–º
directors_stats = filtered_df.groupby('director_first').agg(
    total_support=('refundable_support', 'sum'),  # –°—É–º–º–∞ –≤–æ–∑–≤—Ä–∞—Ç–Ω–æ–π –ø–æ–¥–¥–µ—Ä–∂–∫–∏
    total_nonrefundable_support=('nonrefundable_support', 'sum'),  # –°—É–º–º–∞ –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–Ω–æ–π –ø–æ–¥–¥–µ—Ä–∂–∫–∏
    total_box_office=('box_office', 'sum'),  # –°—É–º–º–∞ —Å–±–æ—Ä–æ–≤
    avg_ros=('ros', 'mean')  # –°—Ä–µ–¥–Ω–∏–π ROS
).reset_index()
# –°—á–∏—Ç–∞–µ–º –æ–±—â—É—é –ø–æ–¥–¥–µ—Ä–∂–∫—É (–≤–æ–∑–≤—Ä–∞—Ç–Ω–∞—è + –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–Ω–∞—è)
directors_stats['total_support'] = (
    directors_stats['total_support'].fillna(0) + 
    directors_stats['total_nonrefundable_support'].fillna(0)
)
```

```python
# –£–±–∏—Ä–∞–µ–º –Ω–µ–Ω—É–∂–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏
directors_stats = directors_stats[['director_first', 'total_support', 'total_box_office', 'avg_ros']]
top_directors = directors_stats.sort_values(by='total_box_office', ascending=False).head(10)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —á–∏—Å–ª–∞ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–µ–ª–∏–º –Ω–∞ –º–∏–ª–ª–∏–æ–Ω—ã)
top_directors['total_support'] = (top_directors['total_support'] / 1e6).round(2)
top_directors['total_box_office'] = (top_directors['total_box_office'] / 1e6).round(2)
top_directors['avg_ros'] = top_directors['avg_ros'].round(2)
```

```python
top_directors.rename(columns={
    'total_support': '–ì–æ—Å. –ø–æ–¥–¥–µ—Ä–∂–∫–∞ (–º–ª–Ω)',
    'total_box_office': '–°–±–æ—Ä—ã (–º–ª–Ω)',
    'avg_ros': '–°—Ä–µ–¥–Ω–∏–π ROS'
}, inplace=True)
top_directors
```
